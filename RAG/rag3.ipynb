```python
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
documents = [
    "JWT token'ları modern API'lerde kimlik doğrulama için kullanılır.",
    "OAuth2 üçüncü parti uygulamalar için yetkilendirme sağlar.",
    "JWT token'ları local storage veya cookie içinde saklanabilir.",
    "OAuth2 authorization code ve client credentials akışlarını destekler.",
    "Authentication kullanıcıyı doğrular, authorization ise kullanıcının yetkilerini belirler."
]
```

```python
model=SentenceTransformer("all-MiniLM-L6-v2")
```

```python
model
```

```python
doc_embeddings=model.encode(documents)
```

```python
doc_embeddings
```

```python
query="JWT autehntıcation nasıl çalışır"
```

```python
query_embedding=model.encode(query)
```

```python
query_embedding
```

```python
scores=np.dot(doc_embeddings,query_embedding.T).flatten()
```

```python
best_index=np.argmax(scores)
```

```python
query
```

```python
documents[best_index]
```

```python
bad_chunks = [
    "JWT token'ları kullanılır.",
    "OAuth2 yetkilendirme sağlar.",
    "Authentication ve authorization farklıdır.",
    "Token'lar cookie içinde tutulabilir."
]
```

```python
chunk_embeddings=model.encode(bad_chunks)
```

```python
scores=np.dot(chunk_embeddings,query_embedding.T).flatten()
chunk_index=np.argmax(scores)
```

```python
bad_chunks[chunk_index]
```

```python
query
```

```python
semantic_scores=np.dot(doc_embeddings,query_embedding.T).flatten()
```

```python
semantic_scores
```

```python
semantic_index=np.argmax(semantic_scores)
```

```python
documents[semantic_index]
```

```python
vectorizer=TfidfVectorizer()
```

```python
tfidf_matrix=vectorizer.fit_transform(documents)
```

```python
query_matrix=vectorizer.transform([query])
```

```python
keyword_scores=np.dot(tfidf_matrix.toarray(),query_matrix.toarray().T).flatten()
```

```python
keyword_index=np.argmax(keyword_scores)
```

```python
documents[keyword_index]
```

```python
context=documents[semantic_index]
```

```python
context
```
